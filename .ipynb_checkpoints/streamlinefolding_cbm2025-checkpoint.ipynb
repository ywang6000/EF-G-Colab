```python
# searching for best pLDDT > threshold, then move those into sel_fold folder. 
import os
import shutil
import csv
import re

# Define the log file path, CSV path, PDB directories, and threshold
log_file_path = '/home/yuhong/localcolabfold/SEQ/Cluster/Fold/log.txt'  # Replace with the actual path
csv_output_path = '/home/yuhong/localcolabfold/SEQ/Cluster/Fold/pLDDT.csv'  # Path for the output CSV file
pLDDT_threshold = 70.0  # Set your threshold for pLDDT
pdb_source_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/Fold'  # Directory containing the PDB files
pdb_target_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb'  # Directory to move selected PDB files

# Ensure the target directory exists
os.makedirs(pdb_target_dir, exist_ok=True)

# Function to process the log file and extract pLDDT values and EX_xxx codes
def process_log_file_and_generate_csv(log_file_path, csv_output_path, threshold):
    with open(log_file_path, 'r') as log_file:
        lines = log_file.readlines()

    extracted_data = []

    # Process the lines in the log file
    for i, line in enumerate(lines):
        # Look for lines containing "rank_001"
        if "rank_001" in line:
            # Extract the pLDDT value
            pLDDT_match = re.search(r'pLDDT=([\d.]+)', line)
            if pLDDT_match:
                pLDDT_value = float(pLDDT_match.group(1))
                print(f"Extracted pLDDT: {pLDDT_value}")  # Debugging

                # Search backward for the corresponding Query line
                for j in range(i, -1, -1):
                    if "Query" in lines[j]:
                        # Extract the EX_xxx string with exactly 6 characters
                        ex_match = re.search(r'(EX_\w{3})', lines[j])
                        if ex_match:
                            ex_code = ex_match.group(0)
                            print(f"Found EX code: {ex_code}")  # Debugging
                            extracted_data.append([ex_code, pLDDT_value])
                            break

    # Write the extracted data to CSV
    with open(csv_output_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(['EX_Code', 'pLDDT_Value'])  # Header
        writer.writerows(extracted_data)

    return extracted_data

# Function to move PDB files if pLDDT is above the threshold and rename them
def move_and_rename_pdb_files(extracted_data, source_dir, target_dir, threshold):
    for ex_code, pLDDT_value in extracted_data:
        if pLDDT_value > threshold:
            # Search for files that start with EX_xxx_unrelaxed_rank_001
            for file_name in os.listdir(source_dir):
                if file_name.startswith(f'{ex_code}_unrelaxed_rank_001'):
                    pdb_source_path = os.path.join(source_dir, file_name)
                    new_file_name = f'{ex_code}.pdb'  # Truncated name
                    pdb_target_path = os.path.join(target_dir, new_file_name)

                    # Move the file to the target directory and rename it
                    shutil.move(pdb_source_path, pdb_target_path)
                    print(f'Moved and renamed: {file_name} -> {new_file_name}')
                    break
            else:
                print(f'File not found for EX code: {ex_code}')

# Process the log file and generate the CSV
extracted_data = process_log_file_and_generate_csv(log_file_path, csv_output_path, pLDDT_threshold)

# Move and rename the PDB files
move_and_rename_pdb_files(extracted_data, pdb_source_dir, pdb_target_dir, pLDDT_threshold)

print("PDB file processing complete.")
```

```python
# Load all selected pdb into pymol, then align them to the highest confident structure.
# Load all selected pdb into pymol, then align them to the highest confident structure.
import os
import shutil
import pandas as pd
import numpy as np
import pymol2

# Define paths
pdb_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/'        # Replace with your PDB files directory
csv_file_path = '/home/yuhong/localcolabfold/SEQ/Cluster/Fold/pLDDT.csv'   # Replace with the path to your CSV file

# Step 1: Move the CSV file into the PDB files directory
csv_file_name = os.path.basename(csv_file_path)
destination_csv_path = os.path.join(pdb_dir, csv_file_name)
shutil.move(csv_file_path, destination_csv_path)

print(f"CSV file moved to: {destination_csv_path}")

# Step 2: Read and rearrange the CSV file
# Read the CSV file into a DataFrame
df = pd.read_csv(destination_csv_path)

# Step 3: Sort the DataFrame based on the second column in descending order
# Assuming the second column is at index 1 (zero-based indexing)
second_column_name = df.columns[1]
sorted_df = df.sort_values(by=second_column_name, ascending=False)

# Step 4: Rewrite the CSV file with the sorted data
sorted_df.to_csv(destination_csv_path, index=False)

print(f"CSV file sorted by '{second_column_name}' in descending order and saved.")

# Step 5: Proceed to align the PDB files
# Get list of PDB files
pdb_files = [f for f in os.listdir(pdb_dir) if f.endswith('.pdb')]

# Use the first PDB file as the reference
reference_pdb = pdb_files[0]
reference_name = os.path.splitext(reference_pdb)[0]

with pymol2.PyMOL() as pymol:
    cmd = pymol.cmd

    # Load the reference PDB
    cmd.load(os.path.join(pdb_dir, reference_pdb), reference_name)

    # List to store object names
    object_names = [reference_name]

    for pdb_file in pdb_files:
        target_name = os.path.splitext(pdb_file)[0]

        if pdb_file == reference_pdb:
            continue  # Skip the reference file

        # Load the target PDB
        cmd.load(os.path.join(pdb_dir, pdb_file), target_name)

        # Align the target to the reference
        cmd.align(target_name, reference_name)

        # Save the aligned structure, overwriting the original file
        cmd.save(os.path.join(pdb_dir, pdb_file), target_name)

        # Keep the target object in the session for RMSD calculation
        object_names.append(target_name)

    # Optional: Adjust visualization settings here if needed

    # Save the session as 'Align_all.pse' in the PDB directory
    session_path = os.path.join(pdb_dir, 'Align_all.pse')
    cmd.save(session_path)

    print(f"Session saved as: {session_path}")

    # Step 6: Compute the RMSD matrix
    num_objects = len(object_names)
    rmsd_matrix = np.zeros((num_objects, num_objects))

    for i in range(num_objects):
        obj1 = object_names[i]
        for j in range(i, num_objects):
            obj2 = object_names[j]

            # Since the structures are already aligned, use 'rms_cur' without further alignment
            rmsd = cmd.rms_cur(obj1, obj2, quiet=1)
            rmsd_matrix[i, j] = rmsd
            rmsd_matrix[j, i] = rmsd  # Symmetric matrix

    # Create DataFrame with headers
    df_rmsd = pd.DataFrame(rmsd_matrix, index=object_names, columns=object_names)

    # Save the RMSD matrix to CSV
    rmsd_csv_path = os.path.join(pdb_dir, 'RMSD_matrix.csv')
    df_rmsd.to_csv(rmsd_csv_path)

    print(f"RMSD matrix saved to: {rmsd_csv_path}")

print("All PDB files have been aligned, saved, and the session has been saved as 'Align_all.pse'.")
```

```python
# Load all selected pdb into pymol, then align them to the highest confident structure.
# make the RMSD matric and plit t-sne plot and DBscan. 
import os
import shutil
import pandas as pd
import numpy as np
import pymol2
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# Define paths
pdb_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/'        # Replace with your PDB files directory
csv_file_path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/pLDDT.csv'   # Replace with the path to your CSV file

# Step 1: Move the CSV file into the PDB files directory
csv_file_name = os.path.basename(csv_file_path)
destination_csv_path = os.path.join(pdb_dir, csv_file_name)
shutil.move(csv_file_path, destination_csv_path)

print(f"CSV file moved to: {destination_csv_path}")

# Step 2: Read and rearrange the CSV file
df = pd.read_csv(destination_csv_path)

# Step 3: Sort the DataFrame based on the second column in descending order
second_column_name = df.columns[1]
sorted_df = df.sort_values(by=second_column_name, ascending=False)

# Step 4: Rewrite the CSV file with the sorted data
sorted_df.to_csv(destination_csv_path, index=False)
print(f"CSV file sorted by '{second_column_name}' in descending order and saved.")

# Step 5: Proceed to align the PDB files
pdb_files = [f for f in os.listdir(pdb_dir) if f.endswith('.pdb')]

# Use the first PDB file as the reference
reference_pdb = pdb_files[0]
reference_name = os.path.splitext(reference_pdb)[0]

with pymol2.PyMOL() as pymol:
    cmd = pymol.cmd

    # Load the reference PDB
    cmd.load(os.path.join(pdb_dir, reference_pdb), reference_name)

    # List to store object names
    object_names = [reference_name]

    for pdb_file in pdb_files:
        target_name = os.path.splitext(pdb_file)[0]

        if pdb_file == reference_pdb:
            continue  # Skip the reference file

        # Load the target PDB
        cmd.load(os.path.join(pdb_dir, pdb_file), target_name)

        # Align the target to the reference
        cmd.align(target_name, reference_name)

        # Save the aligned structure, overwriting the original file
        cmd.save(os.path.join(pdb_dir, pdb_file), target_name)

        # Keep the target object in the session for RMSD calculation
        object_names.append(target_name)

    # Save the session as 'Align_all.pse' in the PDB directory
    session_path = os.path.join(pdb_dir, 'Align_all.pse')
    cmd.save(session_path)
    print(f"Session saved as: {session_path}")

    # Step 6: Compute the RMSD matrix
    num_objects = len(object_names)
    rmsd_matrix = np.zeros((num_objects, num_objects))

    for i in range(num_objects):
        obj1 = object_names[i]
        for j in range(i, num_objects):
            obj2 = object_names[j]

            # Since the structures are already aligned, use 'rms_cur' without further alignment
            rmsd = cmd.rms_cur(obj1, obj2, quiet=1)
            rmsd_matrix[i, j] = rmsd
            rmsd_matrix[j, i] = rmsd  # Symmetric matrix

    # Create DataFrame with headers
    df_rmsd = pd.DataFrame(rmsd_matrix, index=object_names, columns=object_names)

    # Save the RMSD matrix to CSV
    rmsd_csv_path = os.path.join(pdb_dir, 'RMSD_matrix.csv')
    df_rmsd.to_csv(rmsd_csv_path)
    print(f"RMSD matrix saved to: {rmsd_csv_path}")

print("All PDB files have been aligned, saved, and the session has been saved as 'Align_all.pse'.")

# Step 7: Perform t-SNE and plot the results

# Read the RMSD matrix
df_rmsd = pd.read_csv(rmsd_csv_path, index_col=0)
object_names = df_rmsd.index.tolist()

# Convert the RMSD matrix into a NumPy array for t-SNE
rmsd_matrix = df_rmsd.values

# Adjust perplexity to be less than the number of PDB files
n_samples = rmsd_matrix.shape[0]  # Number of samples (PDB files)
perplexity_value = min(30, n_samples - 1)  # Set perplexity to a smaller value

# Initialize t-SNE with the adjusted perplexity and init as "random"
tsne = TSNE(n_components=2, metric='precomputed', perplexity=perplexity_value, init="random", random_state=42)

# Perform t-SNE embedding
tsne_results = tsne.fit_transform(rmsd_matrix)

# Create a DataFrame to store t-SNE results and labels
tsne_df = pd.DataFrame({
    'tsne-2d-one': tsne_results[:, 0],
    'tsne-2d-two': tsne_results[:, 1],
    'label': object_names
})

# Plotting the t-SNE results with a reasonable figure size
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='tsne-2d-one', y='tsne-2d-two',
    hue='label',
    palette=sns.color_palette("hsv", len(object_names)),
    data=tsne_df,
    legend="full",
    alpha=0.8
)
plt.title("t-SNE projection of RMSD matrix")
plt.show()

# Annotate each point with the label
for i in range(tsne_df.shape[0]):
    plt.text(
        tsne_df['tsne-2d-one'][i],
        tsne_df['tsne-2d-two'][i],
        tsne_df['label'][i],
        fontsize=9
    )

plt.title('t-SNE Visualization of PDB Structures Based on RMSD')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.xlim(min(tsne_df['tsne-2d-one']) - 10, max(tsne_df['tsne-2d-one']) + 10)
plt.ylim(min(tsne_df['tsne-2d-two']) - 10, max(tsne_df['tsne-2d-two']) + 10)
plt.show()
```

```python
# reads the 2D RMSD matrix, sorts it to group together all PDB structures within an RMSD threshold of 3.1, 
#and extracts submatrices iteratively until no further grouping is possible.

import pandas as pd
import numpy as np

# Load the 2D RMSD matrix
rmsd_csv_path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/RMSD_matrix.csv'  # Replace with your actual path
df_rmsd = pd.read_csv(rmsd_csv_path, index_col=0)
rmsd_matrix = df_rmsd.values

# Define the threshold for resemblance (RMSD < 3.1)
threshold = 2.8

# List to store clusters
clusters = []

# Function to pool all IDs within a given threshold from the first ID in a group
def pool_ids(df_rmsd, starting_id, threshold):
    pool = set([starting_id])  # Initialize the pool with the starting ID
    to_check = [starting_id]  # List of IDs to check for further pooling

    while to_check:
        current_id = to_check.pop(0)
        current_idx = df_rmsd.index.get_loc(current_id)
        
        # Iterate through remaining IDs to find those within the threshold
        for other_id in df_rmsd.index:
            if other_id not in pool and df_rmsd.loc[current_id, other_id] <= threshold:
                pool.add(other_id)
                to_check.append(other_id)
    
    return list(pool)

# Process the RMSD matrix and group the IDs iteratively
while len(df_rmsd) > 0:
    first_id = df_rmsd.index[0]
    
    # Pool all IDs within the threshold from the first ID
    current_cluster = pool_ids(df_rmsd, first_id, threshold)
    clusters.append(current_cluster)
    
    # Remove the pooled IDs from the RMSD matrix
    df_rmsd = df_rmsd.drop(current_cluster, axis=0).drop(current_cluster, axis=1)
    
    print(f"Cluster created with IDs: {current_cluster}")

# Save the clusters to a text file
cluster_output_path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/cluster_list.txt'
with open(cluster_output_path, 'w') as f:
    for idx, cluster in enumerate(clusters):
        f.write(f"Cluster {idx+1}: {', '.join(cluster)}\n")

print(f"Clusters saved to {cluster_output_path}")

```

```python
# combine a3m based on structure clustering
import os
import shutil

# Define the paths
cluster_txt_path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/cluster_list.txt'  # Path to cluster_list.txt
a3m_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/'  # Directory where all the .a3m files are located
output_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/'  # Output directory where combined .a3m files will be saved

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to combine .a3m files
def combine_a3m_files(cluster_num, a3m_files, output_dir):
    combined_content = []
    first_file = a3m_files[0]
    
    # Read the first file and prefix C# to all headers
    with open(first_file, 'r') as f:
        lines = f.readlines()
        for i in range(0, len(lines), 2):
            header = lines[i].strip()
            sequence = lines[i+1].strip()
            
            # Modify the header with cluster prefix
            new_header = f">C{cluster_num}_" + header[1:]
            combined_content.append(new_header + '\n')
            combined_content.append(sequence + '\n')
    
    # For subsequent files, skip the first sequence and add the rest
    for a3m_file in a3m_files[1:]:
        with open(a3m_file, 'r') as f:
            lines = f.readlines()
        
        # Skip the first sequence (>xxx and its sequence)
        for i in range(2, len(lines), 2):
            header = lines[i].strip()
            sequence = lines[i+1].strip()
            
            # Modify the header with cluster prefix
            new_header = f">C{cluster_num}_" + header[1:]
            combined_content.append(new_header + '\n')
            combined_content.append(sequence + '\n')
    
    # Write the combined file with the cluster name only
    output_filename = f"C{cluster_num}.a3m"
    output_filepath = os.path.join(output_dir, output_filename)
    
    with open(output_filepath, 'w') as out_f:
        out_f.writelines(combined_content)
    
    print(f"Combined file saved as: {output_filepath}")

# Read the cluster_list.txt file
with open(cluster_txt_path, 'r') as f:
    clusters = f.readlines()

# Process each cluster
for cluster in clusters:
    # Extract cluster information
    cluster_info = cluster.strip().split(':')
    cluster_num = int(cluster_info[0].replace('Cluster ', '').strip())
    ids = [id_.strip() for id_ in cluster_info[1].split(',')]
    
    # Find corresponding .a3m files in the a3m_dir
    a3m_files = [os.path.join(a3m_dir, id_ + '.a3m') for id_ in ids if os.path.exists(os.path.join(a3m_dir, id_ + '.a3m'))]
    
    if len(a3m_files) > 1:
        # If the cluster has more than 1 item, combine the .a3m files
        combine_a3m_files(cluster_num, a3m_files, output_dir)
    elif len(a3m_files) == 1:
        # If the cluster has only 1 item, just move the .a3m file to the output directory
        shutil.move(a3m_files[0], output_dir)
        print(f"Moved {os.path.basename(a3m_files[0])} to {output_dir}")
    else:
        print(f"No .a3m files found for Cluster {cluster_num}")

print("Processing complete.")
```

```python
# this cell combines structure clustered a3m. and combine them with cluster tag. 
import os

# Define the paths
cluster_txt_path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/cluster_list.txt'  # Path to cluster_list.txt
a3m_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/'  # Directory where all the .a3m files are located
output_dir = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/'  # Output directory where combined .a3m files will be saved
concatenated_fasta_path = os.path.join(output_dir, 'Concatenated_All.fasta')  # Save concatenated .a3m files as a single fasta

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Dictionary to store line counts for each C#.a3m file
c_file_line_counts = {}

# Function to combine or modify .a3m files and append to the grand fasta
def combine_a3m_files(cluster_num, a3m_files, output_dir):
    combined_content = []
    first_file = a3m_files[0]
    
    # Read the first file and prefix C# to all headers
    with open(first_file, 'r') as f:
        lines = f.readlines()
        for i in range(0, len(lines), 2):
            header = lines[i].strip()
            sequence = lines[i+1].strip()
            
            # Modify the header with cluster prefix
            new_header = f">C{cluster_num}_" + header[1:]
            combined_content.append(new_header)
            combined_content.append(sequence)
    
    # For subsequent files, skip the first sequence and add the rest
    for a3m_file in a3m_files[1:]:
        with open(a3m_file, 'r') as f:
            lines = f.readlines()
        
        # Skip the first sequence (>xxx and its sequence)
        for i in range(2, len(lines), 2):
            header = lines[i].strip()
            sequence = lines[i+1].strip()
            
            # Modify the header with cluster prefix
            new_header = f">C{cluster_num}_" + header[1:]
            combined_content.append(new_header)
            combined_content.append(sequence)
    
    # Write the combined file with the cluster name only
    output_filename = f"C{cluster_num}.a3m"
    output_filepath = os.path.join(output_dir, output_filename)
    
    with open(output_filepath, 'w') as out_f:
        out_f.writelines(f"{line}\n" for line in combined_content)
    
    # Store the line count for this combined file
    c_file_line_counts[output_filename] = len(combined_content)
    
    print(f"Combined file saved as: {output_filepath} with {len(combined_content)} lines")

# Function to modify a single .a3m file by adding the prefix to the headers
def modify_single_a3m(cluster_num, a3m_file, output_dir):
    modified_content = []
    
    with open(a3m_file, 'r') as f:
        lines = f.readlines()
        for i in range(0, len(lines), 2):
            header = lines[i].strip()
            sequence = lines[i+1].strip()
            
            # Modify the header with cluster prefix
            new_header = f">C{cluster_num}_" + header[1:]
            modified_content.append(new_header)
            modified_content.append(sequence)
    
    # Write the modified file with the cluster name
    output_filename = f"C{cluster_num}.a3m"
    output_filepath = os.path.join(output_dir, output_filename)
    
    with open(output_filepath, 'w') as out_f:
        out_f.writelines(f"{line}\n" for line in modified_content)
    
    # Store the line count for this single file
    c_file_line_counts[output_filename] = len(modified_content)
    
    print(f"Modified single file saved as: {output_filepath} with {len(modified_content)} lines")

# Read the cluster_list.txt file
with open(cluster_txt_path, 'r') as f:
    clusters = f.readlines()

# Process each cluster
for cluster in clusters:
    # Extract cluster information
    cluster_info = cluster.strip().split(':')
    cluster_num = int(cluster_info[0].replace('Cluster ', '').strip())
    ids = [id_.strip() for id_ in cluster_info[1].split(',')]
    
    # Find corresponding .a3m files in the a3m_dir
    a3m_files = [os.path.join(a3m_dir, id_ + '.a3m') for id_ in ids if os.path.exists(os.path.join(a3m_dir, id_ + '.a3m'))]
    
    # Combine only Cluster 1 and Cluster 2
    if cluster_num == 1 or cluster_num == 2:
        if len(a3m_files) > 1:
            combine_a3m_files(cluster_num, a3m_files, output_dir)
        elif len(a3m_files) == 1:
            modify_single_a3m(cluster_num, a3m_files[0], output_dir)
    else:
        # Modify the .a3m file by adding the prefix for all other clusters
        if len(a3m_files) == 1:
            modify_single_a3m(cluster_num, a3m_files[0], output_dir)

# Function to filter out sequences with "_101" in the header, except for C1_101
def filter_sequences(content):
    filtered_content = []
    first_c1_101_found = False
    for i in range(0, len(content), 2):
        header = content[i].strip()
        if "_101" in header:
            if "C1_101" in header and not first_c1_101_found:
                first_c1_101_found = True
            else:
                continue  # Skip sequences with "_101" in the header, except "C1_101"
        filtered_content.append(header)
        filtered_content.append(content[i+1])  # Add the sequence
    return filtered_content

# Concatenate all individual .a3m files into one combined fasta file
concatenated_content = []
for c_file in sorted(c_file_line_counts.keys()):
    with open(os.path.join(output_dir, c_file), 'r') as f:
        concatenated_content.extend(f.readlines())

# Filter the concatenated content
filtered_concatenated_content = filter_sequences(concatenated_content)

# Save the filtered concatenated content as a fasta file in the same directory as the .txt file
with open(concatenated_fasta_path, 'w') as concatenated_fasta:
    concatenated_fasta.writelines(f"{line.strip()}\n" for line in filtered_concatenated_content)

print(f"Concatenated and filtered fasta file saved as: {concatenated_fasta_path}")
```

```python
import requests
from Bio import SeqIO
import os
import glob

# Function to fetch protein and organism information from UniProt, if the ID has
# no hit in uniprot, no save in the ourput for downstream analysis. 
def fetch_uniprot_data(uniref_id):
    url = f"https://rest.uniprot.org/uniprotkb/stream?query={uniref_id}&format=tsv&fields=protein_name,organism_name"
    headers = {"User-Agent": "Python script"}
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        lines = response.text.splitlines()
        if len(lines) > 1:
            data = lines[1].split("\t")
            if len(data) >= 2:
                protein_name = data[0]
                organism_name = data[1]
                print(f"Fetched data for {uniref_id}: {protein_name}, {organism_name}")
                return protein_name, organism_name
    else:
        print(f"Failed to fetch data for {uniref_id}. Status code: {response.status_code}")
        print("Response text:")
        print(response.text)
        
    print(f"No data found for {uniref_id}.")
    return None, None

# Function to update FASTA labels and exclude entries without UniProt data
def update_fasta_labels(input_file):
    print(f"Processing {input_file}...")
    records = list(SeqIO.parse(input_file, "fasta"))
    updated_records = []
    
    # Modify the description to replace underscores with dashes and fetch UniProt data
    for record in records:
        original_description = record.description
        
        # Look for UniRef100 ID and extract it
        if "UniRef100_" in original_description:
            uniref_id_part = original_description.split("UniRef100_")[-1].split()[0].strip()  # Extract the UniRef ID part
            
            # Fetch data from UniProt
            protein_name, organism_name = fetch_uniprot_data(uniref_id_part)
            
            # Only add records with valid UniProt data
            if protein_name and organism_name:
                record.description = f"{protein_name} [{organism_name}] | {record.description}"
                updated_records.append(record)
        else:
            print(f"No UniRef100 ID found in {record.id}, skipping.")
    
    # Write the updated records to the output file if there are any
    if updated_records:
        output_file = input_file.replace('.fasta', '_TAX.fa')
        SeqIO.write(updated_records, output_file, "fasta")
        print(f"Updated FASTA written to {output_file}")
    else:
        print(f"No valid records with UniProt data found in {input_file}.")

# Function to process all .fasta files in a directory
def process_directory(path):
    fasta_files = glob.glob(os.path.join(path, '*.fasta'))
    for file_path in fasta_files:
        update_fasta_labels(file_path)

# Set the directory path containing the .fasta files
path = '/home/yuhong/localcolabfold/SEQ/Cluster/sel_pdb/uniprot_tree'

# Now process all FASTA files in this directory
process_directory(path)
```

```python
# after shorten ID, need to generate scv of three columns: protein, taxon, uniref100 ID
import pandas as pd

def process_fasta_file(fasta_file):
    # Initialize lists to store the extracted information
    ids = []
    species_names = []
    descriptions = []

    # Read the fasta file line by line
    with open(fasta_file, 'r') as file:
        for line in file:
            if line.startswith('>'):
                # Extract the ID (first part before the first space or tab)
                id_part = line.split()[0]
                
                # Extract the species name (string between [])
                if '[' in line and ']' in line:
                    species_name = line[line.find('[')+1 : line.find(']')]
                else:
                    species_name = "Unknown"  # Handle cases where species is missing
                
                # Extract the description (string between the first space/tab and first [)
                if ' ' in line and '[' in line:
                    description = line.split(' ', 1)[1].split('[')[0].strip()
                else:
                    description = "No description"  # Handle cases where description is missing
                
                # Append extracted data to lists
                ids.append(id_part)
                species_names.append(species_name)
                descriptions.append(description)

    # Create a DataFrame
    data = {
        'ID': ids,
        'Species Name': species_names,
        'Description': descriptions
    }
    df = pd.DataFrame(data)

    # Generate the output CSV filename
    output_file = fasta_file.replace('.fa', '_summary.csv').replace('.fasta', '_summary.csv')

    # Save the DataFrame to a CSV file
    df.to_csv(output_file, index=False)
    print(f"CSV file saved as: {output_file}")

# Example usage
process_fasta_file(input_fa_file)  # Uses input_fa_file from previous cell.
```

```python
# make phylo tree with iqtree2
def run_phylogenetic_tree(fas_file):
    base_name = os.path.splitext(fas_file)[0]
    # Command to run IQ-TREE for phylogenetic tree construction
    command = [
        'iqtree2',
        '-s', fas_file,
        '-m', 'TEST',
        '-bb', '1000',
        '-nt', 'AUTO',
        '--prefix', base_name
    ]
    subprocess.run(command)
    print(f"Phylogenetic tree constructed for {fas_file}")

```

```python
# or make ancestral tree with iqtree2
def run_phylogenetic_and_ancestral(fas_file):
    base_name = os.path.splitext(fas_file)[0]
    # Command to run IQ-TREE for phylogenetic tree construction
    phylo_command = [
        'iqtree2',
        '-s', fas_file,
        '-m', 'TEST',
        '-bb', '1000',
        '-nt', 'AUTO',
        '--prefix', base_name
    ]
    subprocess.run(phylo_command)
    print(f"Phylogenetic tree constructed for {fas_file}")

    # Command to run ancestral sequence reconstruction using the generated tree
    ancestral_command = [
        'iqtree2',
        '-s', fas_file,
        '-t', f"{base_name}.treefile",
        '-nt', 'AUTO',
        '--ancestral',
        '--prefix', f"{base_name}_anc"
    ]
    subprocess.run(ancestral_command)
    print(f"Ancestral sequences reconstructed for {fas_file}")
```

```python
# Define conditions for label background colors
branch_colors_conditions = {
	"C1":	"#a52a2a",	# brown
	"C2":	"#000000",	# black
	"C3":	"#c0c0c0",	# silver
	"C4":	"#008080",	# teal
	"C5":	"#ffd700",	# yellow
	"C6":	"#ff69b4",	# hot pink
	"C7":	"#ffa500",	# orange
	"C8":	"#800080",	# purple
	"C9":	"#0000ff",	# blue
	#"000450":	"#00ff00",	# green
	#"003050":	"#00ffff",	# cyan
	#"000040":	"#ff0000",	# red
    # Add more conditions as needed
}

def generate_itol_branch_colors(tree_file):
    """Generate an iTOL TREE_COLORS dataset for branch colors."""
    tree = Phylo.read(tree_file, 'newick')
    output_filename = os.path.splitext(tree_file)[0] + '_branch_colors.txt'
    
    with open(output_filename, 'w') as out_file:
        out_file.write("TREE_COLORS\n")
        out_file.write("#lines starting with a hash are comments and ignored during parsing\n")
        out_file.write("SEPARATOR SPACE\n")
        out_file.write("DATA\n")
        
        for clade in tree.find_clades():
            if clade.name:
                for keyword, color in branch_colors_conditions.items():
                    if keyword in clade.name:
                        # Write the branch color information, including style and width
                        # Format: NODE_ID branch COLOR normal 1
                        out_file.write(f"{clade.name} branch {color} normal 1\n")
                        break

    print(f"iTOL branch colors dataset generated: {output_filename}")

import os
import glob  # Import the glob module to use glob.glob()

def process_directory(directory):
    """Process all .treefile files in the specified directory to generate iTOL branch colors datasets."""
    # Ensure that you use glob.glob() here to search for all .treefile files
    for tree_file in glob.glob(os.path.join(directory, '*.treefile')):
        print(f"Processing {tree_file}...")
        generate_itol_branch_colors(tree_file)  # Ensure this function is defined elsewhere

# Set the directory path where your .treefile files are located
directory_path = output_dir  # Assigning the same value

# Ensure the directory exists and process the files
if os.path.isdir(directory_path):
    process_directory(directory_path)
else:
    print("Invalid path or directory does not exist.")
```

```python
# test of ID extraction
def extract_ids(nwk_file):
    with open(nwk_file, 'r') as file:
        newick_content = file.read()

    # Regular expression to match taxon names (excluding numbers and entries starting with numbers)
    taxon_pattern = re.compile(r'([^\s,:();]+)')

    # Find all matches
    taxon_ids = re.findall(taxon_pattern, newick_content)

    # Remove duplicates and filter out entries that are purely numeric or start with a digit
    filtered_ids = {taxon_id for taxon_id in set(taxon_ids) if not re.match(r'^\d', taxon_id)}

    # Extract only the last part of the ID after the last hyphen
    final_ids = {taxon_id.split('-')[-1] for taxon_id in filtered_ids}

    return final_ids

# Specify the path to your Newick (.nwk or .treefile) file
nwk_file_path = "./list1list2_TAX_shortened_aligned.treefile"

# Extract IDs from the file
ids = extract_ids(nwk_file_path)

# Generate the output filename based on the input filename
input_filename = os.path.basename(nwk_file_path)
output_filename = f"{os.path.splitext(input_filename)[0]}_IDs.csv"
output_filepath = os.path.join(os.path.dirname(nwk_file_path), output_filename)

# Write the IDs to a CSV file
with open(output_filepath, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Taxon_ID'])  # Write the header
    for taxon_id in sorted(ids):
        writer.writerow([taxon_id])

print(f"Extracted IDs have been written to {output_filepath}")
```

```python
# add another color to compare to trees for common IDs
# Define the color mapping based on common categories between two trees
# Updated category colors for common branches

def extract_id(label):
    """
    Extracts the last part of the label after the last hyphen up to the colon,
    which indicates the start of the branch length.
    Example: 'EFG-list2-UniRef100-A0A7Y4R8I1:0.4184184849' returns 'A0A7Y4R8I1'.
    """
    if ':' in label:
        label = label.split(':')[0]
    if '-' in label:
        label = label.split('-')[-1]
    return label

def compare_trees_and_color_branches(tree_file1, tree_file2, output_suffix):
    # Read both trees
    tree1 = Phylo.read(tree_file1, 'newick')
    tree2 = Phylo.read(tree_file2, 'newick')

    # Extract and process labels from both trees using the updated extract_id
    labels1 = {clade.name: extract_id(clade.name) for clade in tree1.get_terminals() if clade.name}
    labels2 = {extract_id(clade.name) for clade in tree2.get_terminals() if clade.name}

    # Find common labels based on the extracted IDs
    common_labels = set(labels1.values()).intersection(labels2)

    # Generate iTOL dataset for common labels
    dataset_content = 'DATASET_COLORSTRIP\nSEPARATOR SPACE\nDATASET_LABEL common_labels\nCOLOR #ff4500\nDATA\n'
    for label, shortened_id in labels1.items():
        if shortened_id in common_labels:
            dataset_content += f'{label} #4c1130\n'  # Color common labels from tree_file1 red
    # Define the output file path for the iTOL dataset
    output_file_path = f"{directory_path}common_branches_{output_suffix}.txt"
    with open(output_file_path, 'w') as out_file:
        out_file.write(dataset_content)
    print(f"iTOL dataset file created: {output_file_path}")

    # Optionally save the full IDs from tree_file1 that are common to both trees
    common_ids_path = f'common_ids_from_{os.path.basename(tree_file1)}.txt'
    with open(common_ids_path, 'w') as id_file:
        for label, shortened_id in labels1.items():
            if shortened_id in common_labels:
                id_file.write(label + '\n')
    print(f"Common IDs from {tree_file1} saved to {common_ids_path}")

# Example usage (Update these paths to match your files)
tree_file1 = f"{directory_path}/WT_All_aligned_anc.treefile"
tree_file2 = f"{directory_path}/Node10list1list2_TAX_shortened_aligned.treefile"
compare_trees_and_color_branches(tree_file1, tree_file2, "dataset")

```
