```python
# searching for best pLDDT > threshold, then move those into sel_fold folder. 
import os
import shutil
import csv
import re

# Define the log file path, CSV path, PDB directories, and threshold
log_file_path = '/home/yuhong/demo/SEQ/Cluster/Fold/log.txt'  # Replace with the actual path
csv_output_path = '/home/yuhong/demo/SEQ/Cluster/Fold/pLDDT.csv'  # Path for the output CSV file
pLDDT_threshold = 60.0  # Set your threshold for pLDDT
pdb_source_dir = '/home/yuhong/demo/SEQ/Cluster/Fold'  # Directory containing the PDB files
pdb_target_dir = '/home/yuhong/demo/SEQ/Cluster/sel_pdb'  # Directory to move selected PDB files

# Ensure the target directory exists
os.makedirs(pdb_target_dir, exist_ok=True)

# Function to process the log file and extract pLDDT values and EX_xxx codes
def process_log_file_and_generate_csv(log_file_path, csv_output_path, threshold):
    with open(log_file_path, 'r') as log_file:
        lines = log_file.readlines()

    extracted_data = []

    # Process the lines in the log file
    for i, line in enumerate(lines):
        # Look for lines containing "rank_001"
        if "rank_001" in line:
            # Extract the pLDDT value
            pLDDT_match = re.search(r'pLDDT=([\d.]+)', line)
            if pLDDT_match:
                pLDDT_value = float(pLDDT_match.group(1))
                print(f"Extracted pLDDT: {pLDDT_value}")  # Debugging

                # Search backward for the corresponding Query line
                for j in range(i, -1, -1):
                    if "Query" in lines[j]:
                        # Extract the EX_xxx string with exactly 6 characters
                        ex_match = re.search(r'(EX_\w{3})', lines[j])
                        if ex_match:
                            ex_code = ex_match.group(0)
                            print(f"Found EX code: {ex_code}")  # Debugging
                            extracted_data.append([ex_code, pLDDT_value])
                            break

    # Write the extracted data to CSV
    with open(csv_output_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(['EX_Code', 'pLDDT_Value'])  # Header
        writer.writerows(extracted_data)

    return extracted_data

# Function to move PDB files if pLDDT is above the threshold and rename them
def move_and_rename_pdb_files(extracted_data, source_dir, target_dir, threshold):
    for ex_code, pLDDT_value in extracted_data:
        if pLDDT_value > threshold:
            # Search for files that start with EX_xxx_unrelaxed_rank_001
            for file_name in os.listdir(source_dir):
                if file_name.startswith(f'{ex_code}_unrelaxed_rank_001'):
                    pdb_source_path = os.path.join(source_dir, file_name)
                    new_file_name = f'{ex_code}.pdb'  # Truncated name
                    pdb_target_path = os.path.join(target_dir, new_file_name)

                    # Move the file to the target directory and rename it
                    shutil.move(pdb_source_path, pdb_target_path)
                    print(f'Moved and renamed: {file_name} -> {new_file_name}')
                    break
            else:
                print(f'File not found for EX code: {ex_code}')

# Process the log file and generate the CSV
extracted_data = process_log_file_and_generate_csv(log_file_path, csv_output_path, pLDDT_threshold)

# Move and rename the PDB files
move_and_rename_pdb_files(extracted_data, pdb_source_dir, pdb_target_dir, pLDDT_threshold)

print("PDB file processing complete.")
```

```python
# Load all selected pdb into pymol, then align them to the highest confident structure.
# make the RMSD matric and plit t-sne plot and DBscan. 
import os
import shutil
import pandas as pd
import numpy as np
import pymol2
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# Define paths
pdb_dir = '/home/yuhong/demo/SEQ/Cluster/sel_pdb/'        # Replace with your PDB files directory
csv_file_path = '/home/yuhong/demo/SEQ/Cluster/sel_pdb/pLDDT.csv'   # Replace with the path to your CSV file

# Step 1: Move the CSV file into the PDB files directory
csv_file_name = os.path.basename(csv_file_path)
destination_csv_path = os.path.join(pdb_dir, csv_file_name)
shutil.move(csv_file_path, destination_csv_path)

print(f"CSV file moved to: {destination_csv_path}")

# Step 2: Read and rearrange the CSV file
df = pd.read_csv(destination_csv_path)

# Step 3: Sort the DataFrame based on the second column in descending order
second_column_name = df.columns[1]
sorted_df = df.sort_values(by=second_column_name, ascending=False)

# Step 4: Rewrite the CSV file with the sorted data
sorted_df.to_csv(destination_csv_path, index=False)
print(f"CSV file sorted by '{second_column_name}' in descending order and saved.")

# Step 5: Proceed to align the PDB files
pdb_files = [f for f in os.listdir(pdb_dir) if f.endswith('.pdb')]

# Use the first PDB file as the reference
reference_pdb = pdb_files[0]
reference_name = os.path.splitext(reference_pdb)[0]

with pymol2.PyMOL() as pymol:
    cmd = pymol.cmd

    # Load the reference PDB
    cmd.load(os.path.join(pdb_dir, reference_pdb), reference_name)

    # List to store object names
    object_names = [reference_name]

    for pdb_file in pdb_files:
        target_name = os.path.splitext(pdb_file)[0]

        if pdb_file == reference_pdb:
            continue  # Skip the reference file

        # Load the target PDB
        cmd.load(os.path.join(pdb_dir, pdb_file), target_name)

        # Align the target to the reference
        cmd.align(target_name, reference_name)

        # Save the aligned structure, overwriting the original file
        cmd.save(os.path.join(pdb_dir, pdb_file), target_name)

        # Keep the target object in the session for RMSD calculation
        object_names.append(target_name)

    # Save the session as 'Align_all.pse' in the PDB directory
    session_path = os.path.join(pdb_dir, 'Align_all.pse')
    cmd.save(session_path)
    print(f"Session saved as: {session_path}")

    # Step 6: Compute the RMSD matrix
    num_objects = len(object_names)
    rmsd_matrix = np.zeros((num_objects, num_objects))

    for i in range(num_objects):
        obj1 = object_names[i]
        for j in range(i, num_objects):
            obj2 = object_names[j]

            # Since the structures are already aligned, use 'rms_cur' without further alignment
            rmsd = cmd.rms_cur(obj1, obj2, quiet=1)
            rmsd_matrix[i, j] = rmsd
            rmsd_matrix[j, i] = rmsd  # Symmetric matrix

    # Create DataFrame with headers
    df_rmsd = pd.DataFrame(rmsd_matrix, index=object_names, columns=object_names)

    # Save the RMSD matrix to CSV
    rmsd_csv_path = os.path.join(pdb_dir, 'RMSD_matrix.csv')
    df_rmsd.to_csv(rmsd_csv_path)
    print(f"RMSD matrix saved to: {rmsd_csv_path}")

print("All PDB files have been aligned, saved, and the session has been saved as 'Align_all.pse'.")

# Step 7: Perform t-SNE and plot the results

# Read the RMSD matrix
df_rmsd = pd.read_csv(rmsd_csv_path, index_col=0)
object_names = df_rmsd.index.tolist()

# Convert the RMSD matrix into a NumPy array for t-SNE
rmsd_matrix = df_rmsd.values

# Adjust perplexity to be less than the number of PDB files
n_samples = rmsd_matrix.shape[0]  # Number of samples (PDB files)
perplexity_value = min(30, n_samples - 1)  # Set perplexity to a smaller value

# Initialize t-SNE with the adjusted perplexity and init as "random"
tsne = TSNE(n_components=2, metric='precomputed', perplexity=perplexity_value, init="random", random_state=42)

# Perform t-SNE embedding
tsne_results = tsne.fit_transform(rmsd_matrix)

# Create a DataFrame to store t-SNE results and labels
tsne_df = pd.DataFrame({
    'tsne-2d-one': tsne_results[:, 0],
    'tsne-2d-two': tsne_results[:, 1],
    'label': object_names
})

# Plotting the t-SNE results with a reasonable figure size
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='tsne-2d-one', y='tsne-2d-two',
    hue='label',
    palette=sns.color_palette("hsv", len(object_names)),
    data=tsne_df,
    legend="full",
    alpha=0.8
)
plt.title("t-SNE projection of RMSD matrix")
plt.show()

# Annotate each point with the label
for i in range(tsne_df.shape[0]):
    plt.text(
        tsne_df['tsne-2d-one'][i],
        tsne_df['tsne-2d-two'][i],
        tsne_df['label'][i],
        fontsize=9
    )

plt.title('t-SNE Visualization of PDB Structures Based on RMSD')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.xlim(min(tsne_df['tsne-2d-one']) - 10, max(tsne_df['tsne-2d-one']) + 10)
plt.ylim(min(tsne_df['tsne-2d-two']) - 10, max(tsne_df['tsne-2d-two']) + 10)
plt.show()
```

```python
# set the WT_list1 and WT_list2 via supervision. 
# save those into the sel_pdb folder. 
# combine .a3m in /home/yuhong/demo/SEQ/Cluster according to these lists 
# and save the combinded taged .a3m in /home/yuhong/demo/SEQ/Cluster/sel_pdb folder.

def combine_a3m_files(list_file_path, a3m_directory):
    """
    Combines sequences from multiple A3M files into a single A3M file and returns the output file path.
    
    Args:
        list_file_path (str): Path to the text file listing the A3M file names.
        a3m_directory (str): Directory containing the A3M files.
    
    Returns:
        str: Path to the combined A3M file.
    """
    # Extract the base name for the list file to prefix sequence IDs
    base_name = os.path.splitext(os.path.basename(list_file_path))[0]

    # Read the list of A3M files
    with open(list_file_path, 'r') as file:
        a3m_files = [line.strip() for line in file if line.strip()]

    # Initialize a list to store combined sequences
    combined_sequences = []
    for index, a3m_filename in enumerate(a3m_files):
        a3m_file_path = os.path.join(a3m_directory, a3m_filename)
        if not os.path.exists(a3m_file_path):
            print(f"Warning: File {a3m_file_path} not found. Skipping.")
            continue

        with open(a3m_file_path, 'r') as file:
            sequences = file.readlines()
            if index == 0:
                combined_sequences.extend(sequences)
            else:
                combined_sequences.extend(sequences[1:])  # Skip the first sequence

    # Modify sequence IDs and write to a new file
    output_file_path = os.path.join(os.path.dirname(list_file_path), base_name + '.a3m')
    with open(output_file_path, 'w') as outfile:
        for line in combined_sequences:
            if line.startswith('>'):
                outfile.write('>' + base_name + '_' + line[1:])
            else:
                outfile.write(line)

    print(f"Combined A3M file created at: {output_file_path}")
    return output_file_path


def create_final_combined_file(list1_path, list2_path, a3m_directory):
    """
    Combines two intermediate combined A3M files into a single WT_All.a3m file.

    Args:
        list1_path (str): Path to the first list file.
        list2_path (str): Path to the second list file.
        a3m_directory (str): Directory containing the A3M files.

    Returns:
        str: Path to the final combined file (WT_All.a3m).
    """
    # Generate intermediate combined files
    output_file1 = combine_a3m_files(list1_path, a3m_directory)
    output_file2 = combine_a3m_files(list2_path, a3m_directory)

    # Define the final output file path in the directory of the list files
    final_output_path = os.path.join(os.path.dirname(list1_path), 'WT_All.fasta')

    # Combine the two intermediate files
    with open(final_output_path, 'w') as final_combined:
        with open(output_file1, 'r') as file1:
            final_combined.writelines(file1.readlines())
        with open(output_file2, 'r') as file2:
            final_combined.writelines(file2.readlines()[2:])  # Skip the first sequence  of the second file

    print(f"Final combined A3M file created: {final_output_path}")
    return final_output_path


# Execution
a3m_directory = '/home/yuhong/demo/SEQ/Cluster'
list1_path = '/home/yuhong/demo/SEQ/Cluster/sel_pdb/WT_list1.txt'
list2_path = '/home/yuhong/demo/SEQ/Cluster/sel_pdb/WT_list2.txt'

# Create WT_All.a3m in the directory of the list files
final_file = create_final_combined_file(list1_path, list2_path, a3m_directory)
```

```python
import requests
from Bio import SeqIO
import os
import glob

# Function to fetch protein and organism information from UniProt, if the ID has
# no hit in uniprot, no save in the ourput for downstream analysis. 
def fetch_uniprot_data(uniref_id):
    url = f"https://rest.uniprot.org/uniprotkb/stream?query={uniref_id}&format=tsv&fields=protein_name,organism_name"
    headers = {"User-Agent": "Python script"}
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        lines = response.text.splitlines()
        if len(lines) > 1:
            data = lines[1].split("\t")
            if len(data) >= 2:
                protein_name = data[0]
                organism_name = data[1]
                print(f"Fetched data for {uniref_id}: {protein_name}, {organism_name}")
                return protein_name, organism_name
    else:
        print(f"Failed to fetch data for {uniref_id}. Status code: {response.status_code}")
        print("Response text:")
        print(response.text)
        
    print(f"No data found for {uniref_id}.")
    return None, None

# Function to update FASTA labels and exclude entries without UniProt data
def update_fasta_labels(input_file):
    print(f"Processing {input_file}...")
    records = list(SeqIO.parse(input_file, "fasta"))
    updated_records = []
    
    # Modify the description to replace underscores with dashes and fetch UniProt data
    for record in records:
        original_description = record.description
        
        # Look for UniRef100 ID and extract it
        if "UniRef100_" in original_description:
            uniref_id_part = original_description.split("UniRef100_")[-1].split()[0].strip()  # Extract the UniRef ID part
            
            # Fetch data from UniProt
            protein_name, organism_name = fetch_uniprot_data(uniref_id_part)
            
            # Only add records with valid UniProt data
            if protein_name and organism_name:
                record.description = f"{protein_name} [{organism_name}] | {record.description}"
                updated_records.append(record)
        else:
            print(f"No UniRef100 ID found in {record.id}, skipping.")
    
    # Write the updated records to the output file if there are any
    if updated_records:
        output_file = input_file.replace('.fasta', '_TAX.fa')
        SeqIO.write(updated_records, output_file, "fasta")
        print(f"Updated FASTA written to {output_file}")
    else:
        print(f"No valid records with UniProt data found in {input_file}.")

# Function to process all .fasta files in a directory
def process_directory(path):
    fasta_files = glob.glob(os.path.join(path, '*.fasta'))
    for file_path in fasta_files:
        update_fasta_labels(file_path)

# Set the directory path containing the .fasta files
path = '/home/yuhong/demo/SEQ/Cluster/sel_pdb'

# Now process all FASTA files in this directory
process_directory(path)
```

```python
# Define your conditions for renaming here. 
# This is to replace the long ID to shorter ones for itol coloring
import os
import re
from Bio import SeqIO
rename_conditions = {
    #"Drosophila": "umt", 
    "G-like protein ef-g2": "EF-G2",
   "elongation factor g 1": "EF-G1",
    "elongation factor g 2": "EF-G2",
    "tetracycline": "Tetra",
    "gtp-binding protein": "GtpB",
    "p-loop containing": "Ploop",
    "tr-type g": "TrG",
    "elongation factor g": "EFG",
    "elongation factor ef-g": "EFG",
    "elongation factor efg": "EFG",
    "EF-Tu": "Tu",
    "elongation factor Tu": "Tu",
    "EF-2": "EF2",
    "elongation factor 4": "EF4",
    # Add more conditions as needed
}

# Function to rename sequence ID based on conditions
def rename_sequence_id(sequence_id, rename_conditions):
    # Extract the original identifier after the last space, tab, or pipe
    # Split the ID into parts by space, tab, or pipe
    parts = re.split(r'[ \t|]+', sequence_id)
    # Get the last relevant part
    original_id = parts[-1]
    
    # Convert the sequence ID to lowercase for case-insensitive matching
    lower_seq_id = sequence_id.lower()
    
    # Initialize the prefix as empty
    prefix = ""
    
    # Check each condition in the rename_conditions dictionary
    for term, rename in rename_conditions.items():
        if term.lower() in lower_seq_id:
            prefix = rename
            break  # Stop at the first matching condition
    
    # Create the new short name
    if prefix:
        new_id = f"{prefix}-{original_id}"
    else:
        new_id = f"{original_id}"
    
    return new_id
    
    # execute shorten name 
# File path for the input .fa file
input_fa_file = os.path.join(path, 'WT_All_TAX.fa') # Define the file path to Concatenated_All_TAX.fa using the 'path' variable from the previous cell

# Automatically create the output file name by appending "_shortened.fa"
output_fa_file = os.path.splitext(input_fa_file)[0] + "_shortened.fas"

# Parse the input .fa file and rename sequence IDs
with open(output_fa_file, "w") as output_handle:
    for record in SeqIO.parse(input_fa_file, "fasta"):
        original_id = record.description
        new_id = rename_sequence_id(original_id, rename_conditions)
        record.id = new_id
        record.description = ""  # Remove the description after ID
        SeqIO.write(record, output_handle, "fasta")

print(f"Renamed sequences saved to {output_fa_file}")
```

```python
# after shorten ID, need to generate scv of three columns: protein, taxon, uniref100 ID
import pandas as pd

def process_fasta_file(fasta_file):
    # Initialize lists to store the extracted information
    ids = []
    species_names = []
    descriptions = []

    # Read the fasta file line by line
    with open(fasta_file, 'r') as file:
        for line in file:
            if line.startswith('>'):
                # Extract the ID (first part before the first space or tab)
                id_part = line.split()[0]
                
                # Extract the species name (string between [])
                if '[' in line and ']' in line:
                    species_name = line[line.find('[')+1 : line.find(']')]
                else:
                    species_name = "Unknown"  # Handle cases where species is missing
                
                # Extract the description (string between the first space/tab and first [)
                if ' ' in line and '[' in line:
                    description = line.split(' ', 1)[1].split('[')[0].strip()
                else:
                    description = "No description"  # Handle cases where description is missing
                
                # Append extracted data to lists
                ids.append(id_part)
                species_names.append(species_name)
                descriptions.append(description)

    # Create a DataFrame
    data = {
        'ID': ids,
        'Species Name': species_names,
        'Description': descriptions
    }
    df = pd.DataFrame(data)

    # Generate the output CSV filename
    output_file = fasta_file.replace('.fa', '_summary.csv').replace('.fasta', '_summary.csv')

    # Save the DataFrame to a CSV file
    df.to_csv(output_file, index=False)
    print(f"CSV file saved as: {output_file}")

# Example usage
process_fasta_file(input_fa_file)  # Uses input_fa_file from previous cell.
```

```python
# or make ancestral tree with iqtree2
def run_phylogenetic_and_ancestral(fas_file):
    base_name = os.path.splitext(fas_file)[0]
    # Command to run IQ-TREE for phylogenetic tree construction
    phylo_command = [
        'iqtree2',
        '-s', fas_file,
        '-m', 'TEST',
        '-bb', '1000',
        '-nt', 'AUTO',
        '--prefix', base_name
    ]
    subprocess.run(phylo_command)
    print(f"Phylogenetic tree constructed for {fas_file}")

    # Command to run ancestral sequence reconstruction using the generated tree
    ancestral_command = [
        'iqtree2',
        '-s', fas_file,
        '-t', f"{base_name}.treefile",
        '-nt', 'AUTO',
        '--ancestral',
        '--prefix', f"{base_name}_anc"
    ]
    subprocess.run(ancestral_command)
    print(f"Ancestral sequences reconstructed for {fas_file}")
```

```python
# datasets for itol display of protein rings.
# Define three different category_color mappings for proteins
category_colors_A = {
    "EF2": "#ff0000",  # red
    "EFG": "#c0c0c0",  # silver grey
}

category_colors_B = {
    "EF-G2": "#00ffff",  # cyan
    "Tetra": "#0000ff",  # blue
    "EF-G1":"#ffd700", #sunshine yellow
}

category_colors_C = {
    "Ploop": "#800080",  # purple
    "GtpB": "#00ff00",  # green
    "RF2": "#ffa500",  # orange
    "TrG": "#ff69b4",  # hot pink
}

def process_tree_file(tree_file, category_colors, dataset_label, output_suffix):
    tree = Phylo.read(tree_file, 'newick')

    # Note: Changed SEPARATOR to SPACE. Revert to TAB if SPACE is not supported by iTOL.
    dataset_content = (f'DATASET_COLORSTRIP\n'
                       f'SEPARATOR SPACE\n'
                       f'DATASET_LABEL {dataset_label}\n'
                       f'COLOR #ff0000\n'
                       f'DATA\n')

    for label in tree.get_terminals():
        label_name = label.name if label.name else ""
        color = '#ffffff'  # Default color
        for category, category_color in category_colors.items():
            if category in label_name:
                color = category_color
                break
        dataset_content += f'{label.name} {color}\n'

    # Define the output file path
    output_file_path = os.path.splitext(tree_file)[0] + output_suffix + '.txt'
    with open(output_file_path, 'w') as out_file:
        out_file.write(dataset_content)
    print(f"iTOL dataset file created: {output_file_path}")

def process_all_treefiles(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.treefile'):
            tree_file_path = os.path.join(directory, filename)
            print(f"Processing {tree_file_path}...")

            # Process for each category color mapping
            process_tree_file(tree_file_path, category_colors_A, "labelProA", "_itol_datasetA")
            process_tree_file(tree_file_path, category_colors_B, "labelProB", "_itol_datasetB")
            process_tree_file(tree_file_path, category_colors_C, "labelProC", "_itol_datasetC")

# Path to the directory containing .treefile files
directory_path = "."

# Ensure the directory exists
if os.path.isdir(directory_path):
    process_all_treefiles(directory_path)
else:
    print("Invalid path or directory does not exist.")
```

```python
# generate the dataset for leave colors in itol.
# Define conditions for label background colors
branch_colors_conditions = {
    "List1": "#ff0000",  # red
    "List2": "#00ff00",  # green
    "list1": "#ff0000",  # red
    "list2": "#00ff00",  # green
    "-1-": "#FF0000",    # red
    "-2-": "#0000FF",    # blue
    "-3-": "#FFFF00",    # yellow
    "-4-": "#00FF00",    # green
    "-7-": "#FF00FF",    # magenta
    "-8-": "#00FFFF",    # cyan
    "-13-": "#FFA500",   # orange
    "-17-": "#800080",   # purple
    "-24-": "#008000",   # dark green
    "-28-": "#000080",   # navy blue
    "-37-": "#FF4500",   # orange-red
    "-137-": "#8B0000",  # dark red
    "-248-": "#4682B4",  # steel blue
    # Add more conditions as needed
}

def generate_itol_branch_colors(tree_file):
    """Generate an iTOL TREE_COLORS dataset for branch colors."""
    tree = Phylo.read(tree_file, 'newick')
    output_filename = os.path.splitext(tree_file)[0] + '_branch_colors.txt'
    
    with open(output_filename, 'w') as out_file:
        out_file.write("TREE_COLORS\n")
        out_file.write("#lines starting with a hash are comments and ignored during parsing\n")
        out_file.write("SEPARATOR SPACE\n")
        out_file.write("DATA\n")
        
        for clade in tree.find_clades():
            if clade.name:
                for keyword, color in branch_colors_conditions.items():
                    if keyword in clade.name:
                        # Write the branch color information, including style and width
                        # Format: NODE_ID branch COLOR normal 1
                        out_file.write(f"{clade.name} branch {color} normal 1\n")
                        break

    print(f"iTOL branch colors dataset generated: {output_filename}")

import os
import glob  # Import the glob module to use glob.glob()

def process_directory(directory):
    """Process all .treefile files in the specified directory to generate iTOL branch colors datasets."""
    # Ensure that you use glob.glob() here to search for all .treefile files
    for tree_file in glob.glob(os.path.join(directory, '*.treefile')):
        print(f"Processing {tree_file}...")
        generate_itol_branch_colors(tree_file)  # Ensure this function is defined elsewhere

# Set the directory path where your .treefile files are located
directory_path = "."  # Change this to your directory path

# Ensure the directory exists and process the files
if os.path.isdir(directory_path):
    process_directory(directory_path)
else:
    print("Invalid path or directory does not exist.")
```

```python
# extract selection terms, such as EFG, from .a3m.And save as .a3m
# to distinguish the files and for logo generation. 
def extract_sequences_from_a3m(a3m_file, selection_terms, output_fas_file):
    """
    Extract sequences from an A3M file based on selection terms and save them to a new .fasta file.
    The search is non case-sensitive.
    
    Parameters:
    - a3m_file: str, path to the input A3M file.
    - selection_terms: list of str, selection terms to search for in sequence headers.
    - output_fas_file: str, path to the output FASTA file to save pooled sequences.
    
    Returns:
    - None, writes sequences to the output file.
    """
    found_sequences = []
    
    # Convert selection terms to lowercase for non case-sensitive search
    selection_terms_lower = [term.lower() for term in selection_terms]
    
    # Read the input A3M file
    with open(a3m_file, 'r') as file:
        lines = file.readlines()

    # Loop through lines in the file and look for headers and sequences
    current_header = ""
    current_sequence = ""
    
    for line in lines:
        line = line.strip()
        if line.startswith('>'):  # It's a header line
            if current_header and current_sequence:
                # Check if the current header contains any of the selection terms
                current_header_lower = current_header.lower()  # Convert header to lowercase
                for term in selection_terms_lower:
                    if term in current_header_lower:
                        # If the term is found, save the header and sequence
                        found_sequences.append((current_header + " | SelectionTerm: " + term, current_sequence))
                        break  # Stop searching for the next term
            
            # Update the current header and reset the sequence for the next record
            current_header = line
            current_sequence = ""
        else:
            current_sequence += line  # Collect the sequence part

    # Don't forget to check the last record
    if current_header and current_sequence:
        current_header_lower = current_header.lower()  # Convert header to lowercase
        for term in selection_terms_lower:
            if term in current_header_lower:
                found_sequences.append((current_header + " | SelectionTerm: " + term, current_sequence))
                break

    # Write pooled sequences to the output FASTA file
    with open(output_fas_file, 'w') as out_file:
        if found_sequences:
            for header, sequence in found_sequences:
                out_file.write(f"{header}\n{sequence}\n")
        else:
            print(f"No selection term is found in {a3m_file}.")

def process_a3m_files_in_directory(directory_path, selection_terms):
    """
    Process all .a3m files in a directory and extract sequences based on selection terms.
    Outputs a .fasta file for each .a3m file with the selection term appended to the file name.
    
    Parameters:
    - directory_path: str, path to the directory containing .a3m files.
    - selection_terms: list of str, selection terms to search for in sequence headers.
    
    Returns:
    - None
    """
    for file_name in os.listdir(directory_path):
        if file_name.endswith(".a3m"):
            a3m_file_path = os.path.join(directory_path, file_name)
            # Process each selection term and output the corresponding .fas file
            for term in selection_terms:
                output_fas_file = os.path.join(directory_path, file_name.replace(".a3m", f"_{term}.a3m"))
                extract_sequences_from_a3m(a3m_file_path, [term], output_fas_file)

# Define the directory path and selection terms
directory_path = "/home/yuhong/demo/logo/"
#selection_terms = ["RF2", "EF-G2", "EF2","EF-G1", "Tetra", "GtpB","Ploop", "TrG","EFG", "EF2"]
#selection_terms = ["list1", "list2"]
selection_terms = ["EFG"]
# Process all .a3m files in the directory
process_a3m_files_in_directory(directory_path, selection_terms)
print("fine sorting done.")
```

```python
# make logo plots for all .a3m
import os
import logging
from glob import glob
from Bio import AlignIO
import matplotlib.pyplot as plt
import logomaker

# Configure logging
logging.basicConfig(
    filename='process_files.log',
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def create_sequence_logo(aligned_file, output_dir, start_pos, end_pos):
    """Create a sequence logo for a specified range of residues and save it as a PNG file."""
    try:
        alignment = AlignIO.read(aligned_file, "fasta")
        logging.info(f"Successfully read alignment file {aligned_file}")
    except Exception as e:
        logging.error(f"Error reading {aligned_file}: {e}")
        return

    # Ensure sequences are properly aligned
    sequence_length = alignment.get_alignment_length()
    for record in alignment:
        if len(record.seq) < sequence_length:
            record.seq += '-' * (sequence_length - len(record.seq))
        elif len(record.seq) > sequence_length:
            logging.error(f"Sequence {record.id} in {aligned_file} is longer than the alignment length.")
            return

    # Validate positions
    start_pos = max(0, start_pos - 1)  # 0-based indexing correction
    end_pos = min(end_pos, sequence_length)

    # Extract fragments for the specified range
    fragment_alignment = [str(record.seq[start_pos:end_pos]) for record in alignment]

    # Generate logo matrix
    fragment_df = logomaker.alignment_to_matrix(fragment_alignment, to_type='counts', characters_to_ignore='.-')
    fragment_df.index = range(start_pos + 1, end_pos + 1)

    # Create the logo plot
    fig, ax = plt.subplots(figsize=(10, 6))
    logo = logomaker.Logo(fragment_df, ax=ax, color_scheme='skylign_protein')
    ax.set_title(f"Residues {start_pos + 1} to {end_pos} for {os.path.basename(aligned_file)}")
    ax.set_ylabel('Frequency')
    ax.set_xlabel('Position')
    ax.tick_params(axis='x', rotation=90)

    # Save the logo
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        logging.info(f"Created directory {output_dir}")
        
    logo_path = os.path.join(output_dir, f"{os.path.basename(aligned_file)}_logo_{start_pos + 1}_to_{end_pos}.png")
    plt.tight_layout()
    plt.savefig(logo_path)
    plt.close()
    logging.info(f"Saved logo to {logo_path}")


def process_files_in_directory(directory_path, start_pos=1, end_pos=-1, chunk_size=40):
    """Process all .a3m and .fas files in the directory and create sequence logos for fragments."""
    # Collect both .a3m and .fas files
    a3m_files = glob(os.path.join(directory_path, '*.a3m'))
    fas_files = glob(os.path.join(directory_path, '*.fas'))
    
    all_files = a3m_files + fas_files
    logging.info(f"Found {len(all_files)} .a3m and .fas files in {directory_path}")

    if not all_files:
        logging.warning("No .a3m or .fas files found in the specified directory.")
        return

    # Process each file
    for file_path in all_files:
        file_name = os.path.basename(file_path).split('.')[0]
        
        # Create directory for logos
        logo_output_dir = os.path.join(directory_path, 'logos', file_name)
        
        try:
            alignment = AlignIO.read(file_path, "fasta")  # Read as FASTA alignments
            sequence_length = alignment.get_alignment_length()
            end_pos = end_pos if end_pos != -1 else sequence_length

            # Generate logos in chunks
            chunk_start = start_pos
            while chunk_start <= end_pos:
                chunk_end = min(chunk_start + chunk_size - 1, end_pos)
                create_sequence_logo(file_path, logo_output_dir, chunk_start, chunk_end)
                chunk_start += chunk_size
        except Exception as e:
            logging.error(f"Error processing file {file_path}: {e}")


# Execution
directory_path = '/home/yuhong/demo/logo'
start_pos = 1  # Start position (1-based index)
end_pos = -1   # End position (inclusive, set to -1 for end of sequence)
chunk_size = 40  # Size of each fragment

process_files_in_directory(directory_path, start_pos, end_pos, chunk_size)
```

```python
# pairwise comparison of msas, so each collection of msa is unique.

from Bio import SeqIO
import os
import shutil  # Import shutil for file operations

def compare_sequence_files(file1_path, file2_path, output_dir):
    """
    Compare sequences between two files and output unique and common sequences.
    """
    file1_base = os.path.splitext(os.path.basename(file1_path))[0]
    file2_base = os.path.splitext(os.path.basename(file2_path))[0]

    # Parse sequences from both files
    file1_seqs = {record.id: record for record in SeqIO.parse(file1_path, "fasta")}
    file2_seqs = {record.id: record for record in SeqIO.parse(file2_path, "fasta")}

    # Determine common and unique sequences
    common_ids = set(file1_seqs.keys()) & set(file2_seqs.keys())
    unique_file1_ids = set(file1_seqs.keys()) - common_ids
    unique_file2_ids = set(file2_seqs.keys()) - common_ids

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # File paths for output
    output_files = {
        'common': os.path.join(output_dir, f"{file2_base}{file1_base}.fasta"),  # Concatenated name
        'unique1': os.path.join(output_dir, f"{file1_base}.fasta"),  # Updated file1
        'unique2': os.path.join(output_dir, f"{file2_base}0.fasta"),  # Updated file2
    }

    # Write outputs
    for key, seq_ids in [('common', common_ids), ('unique1', unique_file1_ids), ('unique2', unique_file2_ids)]:
        output_path = output_files[key]
        with open(output_path, 'w') as output_handle:
            if key == 'unique1':
                SeqIO.write((file1_seqs[id_] for id_ in seq_ids), output_handle, "fasta")
            elif key == 'unique2':
                SeqIO.write((file2_seqs[id_] for id_ in seq_ids), output_handle, "fasta")
            else:
                SeqIO.write((file1_seqs[id_] for id_ in common_ids), output_handle, "fasta")

        # Remove empty files
        if os.path.getsize(output_path) == 0:
            os.remove(output_path)
            print(f"Deleted empty file: {output_path}")

    # Overwrite file1 with unique sequences for next comparison
    with open(file1_path, 'w') as output_handle:
        SeqIO.write((file1_seqs[id_] for id_ in unique_file1_ids), output_handle, "fasta")

    # Write a summary
    summary_file = os.path.join(output_dir, f"{file1_base}_{file2_base}_summary.txt")
    with open(summary_file, 'w') as f:
        f.write(f"Total sequences in {file1_base}: {len(file1_seqs)}\n")
        f.write(f"Total sequences in {file2_base}: {len(file2_seqs)}\n")
        f.write(f"Common sequences: {len(common_ids)}\n")
        f.write(f"Unique sequences in {file1_base}: {len(unique_file1_ids)}\n")
        f.write(f"Unique sequences in {file2_base}: {len(unique_file2_ids)}\n")

    print(f"Comparison completed for {file1_base} and {file2_base}. Outputs saved to {output_dir}.")

def process_files_in_rounds(file_list, base_folder):
    """
    Process files sequentially, comparing each file with all files in the previous round's folder,
    creating a new folder for each round.
    """
    round_number = 1

    # Use the first file as the initial reference
    initial_file = file_list.pop(0)
    current_round_folder = os.path.join(base_folder, f"Combine_Round{round_number}")
    os.makedirs(current_round_folder, exist_ok=True)
    shutil.copy(initial_file, os.path.join(current_round_folder, os.path.basename(initial_file)))

    while file_list:
        next_file = file_list.pop(0)
        next_file_name = os.path.basename(next_file)
        next_file_path = os.path.join(base_folder, next_file_name)

        # Create the folder for the next round
        next_round_folder = os.path.join(base_folder, f"Combine_Round{round_number + 1}")
        os.makedirs(next_round_folder, exist_ok=True)

        # Compare the new file with all files in the current round's folder
        for fasta_file in os.listdir(current_round_folder):
            if fasta_file.endswith('.fasta'):  # Only process FASTA files
                fasta_path = os.path.join(current_round_folder, fasta_file)
                compare_sequence_files(next_file_path, fasta_path, next_round_folder)

        # Update the new file path after it is overwritten with unique sequences
        next_file_path = os.path.join(next_round_folder, f"{os.path.splitext(next_file_name)[0]}.fasta")

        # Move to the next round
        current_round_folder = next_round_folder
        round_number += 1

# Example Usage
combine_folder = "/home/yuhong/demo/combine_All_MSA"  # Path to the Combine folder
file_list = [
    "/home/yuhong/demo/combine_All_MSA/1.fasta",
    "/home/yuhong/demo/combine_All_MSA/2.fasta",
    "/home/yuhong/demo/combine_All_MSA/3.fasta",
    "/home/yuhong/demo/combine_All_MSA/4.fasta",
    "/home/yuhong/demo/combine_All_MSA/7.fasta",
    "/home/yuhong/demo/combine_All_MSA/8.fasta",
]

process_files_in_rounds(file_list, combine_folder)
```
